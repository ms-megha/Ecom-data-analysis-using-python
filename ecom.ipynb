```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load data
dataset= pd.read_excel("Superstore_USA.xlsx")
dataset.head(2)
```

#check null value

```python
#dataset.isnull().sum()
```

#fill null value

```python
dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean(), inplace=True)
```

#check duplicate

```python
#dataset.duplicated().sum()
```

#analize Order Priority

```python
dataset["Order Priority"].value_counts()
```

```python
dataset["Order Priority"].unique()
```

```python
dataset["Order Priority"]=dataset["Order Priority"].replace("Critical ", "Critical")
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x="Order Priority", data = dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.png")
plt.show()
```

#Ship Mode

```python
dataset["Ship Mode"].value_counts()
x=dataset["Ship Mode"].value_counts().index
y=dataset["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(6,3))
plt.pie(y, labels=x, autopct="%0.2f%%", startangle=30)
plt.legend(loc="best")
plt.title("Ship Mode")
plt.savefig("Count of Ship Mode.png")
plt.show()
```

#product category according to shiping mode

```python
plt.figure(figsize=(6,3))
sns.countplot(x="Ship Mode", data = dataset, hue="Product Category")
plt.title("Count of Ship Mode based on Product Category")
plt.savefig("Count of Ship Mode based on Product Category.png")
plt.show()
```
